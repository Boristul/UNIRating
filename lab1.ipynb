##### Опыт 1.

**Пуля вылетает из ружья на высоте h и приземляется на растоянии L.**

В данном опыте на пулю действует сила тяжести, сила сопротивления воздуха. Но так как ты берем идеальные условия, 
силу сопротивления воздуха мы не учитываем. Тогда выведем формулу для рассчета расстояния полета пули при заданных 
условиях:

**Дано**: 
$V_{0} = 800 m/s$
$m = 0.005 kg$
$g= 9.80665m/s^2$

Исходя из равноускоренного падения пули можно рассчитать время падения пули:
$$h= \sqrt{(gt^2)\over2}$$

$$t= \sqrt{2h\over g}$$

Так как движение по горизонтали прямолинейное, равномерное:
$$L_{0}=V_{0}t=V_{0}\sqrt{2h\over g}$$

На основе этой формулы проанализируем стенд при отклонении на 0.1:

![график 1](ex1.png)

На основе проанализированных данных получили:
$$V_{ovr}=798.06030479101594m/s$$
$$L = 361.4031611621005m$$

Модель верна. Код:

```python
def ex1(h, V0, sigma, n):

    L0 = sqrt((power(V0,2)*h*2)/g)
    L = ([uniform(L0-L0*sigma/2, L0+L0*sigma/2) for x in range(n)])
    V = (sqrt((float_power(L, 2)*g)/(h*2)))
    plot1(V, L)
    return [average(V), L0]
```

##### Опыт 2.

**Cтрельба в ящик с песком на жестком подвесе с измерением угла отклонения.**
В данном опыте необходимо расмотреть стенд относительно 2ух фундаментальных законов физики:
- Закон сохранения энергии
- Закон сохранения импульса

**1.  Закон сохранения энергии**
Запишем ЗСЭ:

$$mV^2/ 2 = (m+M)gl(1-cos(a))$$

Кинетическая энергия пули равна потенциальной энергии полученного объекта.

Высчитаем теоритический угол:
$$\alpha_0 = arccos(\frac{-V_0^2m}{2gl(M+m)}+1)$$

Тогда скорость пули будет равна:
$$V = \sqrt{2gl(1-cos(\alpha))\frac{M+m}{m}}$$

График по отклонению на 0.2
 
 ![график 2.1](ex21.png)
 
 Проанализированная скорость относительно угла:
$$V_{ovr}=796.56131806003464m/s$$
$$a = 0.8315604423745756m$$

Модель верна.

Код:

```python
def ex21(M,m,l,V0,sigma,n):
    alp0 = (1-(m*power(V0,2))/(2*g*l*(M+m)))
    alp0 = arccos(alp0)
    alp = [uniform(alp0-sigma/2*alp0,alp0+sigma/2*alp0)for x in range(n)]
    V= sqrt(2*(M+m)*g*l*(1-cos(alp))/m)
    plot1(V,alp,'V','alp')
    return [average(V), alp0]
```
**2.  Закон сохранения импульса**

Запишем ЗСИ для данной системы:

$$mV_{0} = (m+M)U$$

Формулы для подсчета скорости и угла исходя из ЗСИ и ЗСЭ:

$$U = \frac{M+m}{m}\sqrt{2gl(1-cos(\alpha))}$$
$$\alpha_0=arccos(1-\frac{1}{2gl}(\frac{v_0 m}{(M+m)})^2)$$

График по отклонению на 0.2
 
 ![график 2.2](ex22.png)
 
 Проанализированная скорость относительно угла:
$$V_{ovr}=801.18353969780776m/s$$
$$a = 0.0080776941021158382m$$

Код:
```python
def ex22(M,m,l,V0,sigma,n):
    alp0 = (1-(1/(2*g*l))*power((V0*m)/(M+m),2))
    alp0 = arccos(alp0)
    alp = [uniform(alp0 - sigma / 2 * alp0, alp0 + sigma / 2 * alp0) for x in range(n)]
    V = ((M+m)/m)*sqrt(2*g*l*(1-cos(alp)))
    plot1(V, alp, 'V', 'alp')
    return [average(V), alp0]
```

##### Опыт 3.

На горизонтальной плоскости лежит деревянный брусок,  прикрепленный к вертикальной стенке пружиной.
В центр бруска попадает пуля, летящая горизонтально и параллельно пружине, и застревает в нем.

**Дано**
$$V_{0} = 800 m/s$$
$$m = 0.001 kg$$
$$M = 800 kg$$
$$k = 100 N/m$$

Воспользуемся законом сохранения энергии:
$$(m(V_{0})^2)/2=(k(dl^2))/2$$

В данном случае кинетическая энергия пули переходит в потанциальную энергию пружины.
Из этих формул получим:

$$dl = \frac{Vm}{(\sqrt{(M+m)k)}}$$

$$V=\frac{dl\sqrt{(M+m)k}}{m}$$

График полученных результатов:

![график 3](ex3.png)

$$V_{ovr}=800.04592427770137m/s$$
$$a = 0.079996000299975001m$$

Модель верна.

Код:

```python
def ex3(M,m,V0,k,sigma,n):
    dl = (V0*m)/sqrt((M+m)*k)
    l = [uniform(dl-sigma/2*dl,dl+sigma/2*dl) for x in range(n)]
    V= (power(l,1)*sqrt((M+m)*k))/m
    plot1(V, l, 'V', 'dl')
    return [average(V), dl]
```
